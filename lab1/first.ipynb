```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_iris()
```

```python
iris = pd.DataFrame(data= np.c_[data['data'], data['target']],
                     columns= data['feature_names'] + ['target'])
iris.head() # первые 5 строк
```

```python
iris.shape
```

```python
for x in iris.columns:
    print(x)
```

```python
iris.dtypes
```

```python
for col in iris.columns:
    # Количество пустых значений
    temp_null_count = iris[iris[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
iris.describe()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(ax=ax, x='sepal length (cm)', y='petal length (cm)', data=iris, hue='target')
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(iris['petal length (cm)'])
```
